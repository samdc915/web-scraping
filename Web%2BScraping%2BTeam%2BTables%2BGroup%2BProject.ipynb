```python
soup_table = soup.find_all('p', class_='name')
```

```python
soup_table
```

```python
tables = [pt.get_text() for pt in soup_table]
tables
```

```python
price_tags = soup.find_all('span', class_='price' )
price_tags
```

```python
tables2 = [spant.get_text() for spant in price_tags]
tables2
```

```python
import numpy as np
import pandas as pd
```

```python
table_series = pd.Series(tables2, index=tables)
```

```python
table_series
```

```python
table_frame=table_series.to_frame(name='Price')
table_frame.index.names = ['Table']
```

```python
table_frame
```

```python
with open('sports_tables.csv', 'w', encoding='utf-8') as sports_table:
    table_frame.to_csv(sports_table)
sports_table.close()
```
